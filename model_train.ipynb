```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
loan_stat_df_fet_engineered = pd.read_csv('data/loan_stat_df_for_training.csv',low_memory=False, delimiter=',')
loan_stat_train_df = loan_stat_df_fet_engineered
loan_stat_df_fet_engineered.head()
```

```python
loan_stat_train_df["loan_status_altered"] = loan_stat_train_df["loan_status_altered"].map({"Fully Paid": 0, "Charged Off": 1})
loan_stat_train_df.head()
```

```python
x = np.array(loan_stat_train_df[["loan_amnt","funded_amnt","term","int_rate","installment","emp_length","annual_inc","dti","earliest_cr_line","revol_util","total_pymnt","total_rec_prncp","total_rec_int","recoveries","collection_recovery_fee","last_pymnt_amnt"]])
y = np.array(loan_stat_train_df[["loan_status_altered"]])
```

```python
y = y.ravel()
```

```python
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline

xtrain, xtest, ytrain, ytest = train_test_split(x, y, 
                                                    test_size=0.33, 
                                                    random_state=42)

rf_clf = RandomForestClassifier(criterion='gini', bootstrap=True, random_state=100)
smote_sampler = SMOTE(random_state=9)
pipeline = Pipeline(steps = [['smote', smote_sampler],
                             ['classifier', rf_clf]])
pipeline.fit(xtrain, ytrain)
y_pred = pipeline.predict(xtest)
```

```python
print(y_pred)
```

```python
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, accuracy_score, roc_auc_score
print("-------------------------TEST SCORES-----------------------") 
print(f"Recall: { round(recall_score(ytest, y_pred)*100, 4) }")
print(f"Precision: { round(precision_score(ytest, y_pred)*100, 4) }")
print(f"F1-Score: { round(f1_score(ytest, y_pred)*100, 4) }")
print(f"Accuracy score: { round(accuracy_score(ytest, y_pred)*100, 4) }")
print(f"AUC Score: { round(roc_auc_score(ytest, y_pred)*100, 4) }")
```

```python
cm = confusion_matrix(ytest, y_pred)

# Convert confusion matrix to a pandas DataFrame
cm_df = pd.DataFrame(cm, columns=['Predicted 0', 'Predicted 1'], index=['Actual 0', 'Actual 1'])

# Display confusion matrix as a table
print(cm_df)
```

AUC score:
A model with a higher AUC score indicates that it is better at distinguishing between the two classes, i.e., it has a higher true positive rate while keeping a lower false positive rate. This means that the model is more accurate in identifying loan defaulters while minimizing the risk of falsely classifying non-defaulters as defaulters.

Accuracy: is calculated as the ratio of the number of correctly classified samples to the total number of samples in the dataset. In other words, it measures how often the model is correct in its predictions of both loan defaulters and non-defaulters.

Precision: is the proportion of positive predictions that are correct, while recall is the proportion of actual positive cases that are correctly identified by the model. In other words, precision measures how well the model predicts loan defaulters among all predicted defaulters, while recall measures how well the model identifies all the actual loan defaulters.

high recall: indicates that the model makes few false negative predictions, meaning that the number of cases where the model failed to identify a loan defaulter but the borrower actually defaulted on the loan is low.

The F1 score: combines both precision and recall into a single metric that represents their harmonic mean. A high F1 score indicates that the model has a good balance between precision and recall, and is able to correctly identify loan defaulters while minimizing false positives and false negatives.
